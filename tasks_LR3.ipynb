## 1. Построение простого графика.

Напишите код, который строит график функции y = x2 на интервале [−10, 10]. Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = x**2 

plt.plot(x, y, color="green")

plt.title("График функции y = x^2")
plt.xlabel("x")
plt.ylabel("y")

plt.grid()

plt.show()
```

## 2. Настройка стилей и цветов.

Постройте три линии на одном графике:

1. y = x (синяя, пунктирная линия),
2. y = $x^2$ (зеленая, штрих-пунктирная линия),
3. y = $x^3$ (красная, сплошная линия).

Добавьте легенду и сделайте оси одинакового масштаба.

```python
x = np.linspace(-10, 10, 50)

y1 = x
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

plt.figure(figsize=(5, 5))
plt.plot(x, y1, 'b--', label="y = x")
plt.plot(x, y2, 'g-.', label="y = x^2")
plt.plot(x, y3, 'r-', label="y = x^3")

plt.title("Зависимости: y1 = x, y2 = x^2, y3 = x^3")
plt.xlabel("x")
plt.ylabel("y1, y2, y3")
plt.legend()
plt.grid()

plt.show()
```

## 4. Гистограмма распределения.

Сгенерируйте 1000 случайных чисел из нормального распределения с
параметрами μ = 0, σ = 1 и постройте их гистограмму с 30 бинами. Добавьте
вертикальную линию в среднем значении.

```python
np.random.seed(13)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, alpha=0.5, color='green', edgecolor='black')

mean_value = np.mean(data)
plt.axvline(mean_value, color='black', linestyle='--', linewidth=3, label=f'Среднее = {mean_value:.2f}')

plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.title('Гистограмма нормального распределения (μ=0, σ=1)')
plt.legend()
plt.grid()

plt.show()
```

## 7. Трехмерный график поверхности.

Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = \sin\left(\sqrt{x^2 + y^2}\right)$
на сетке значений x, y в диапазоне [−5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D-график поверхности z = sin(sqrt(x^2 + y^2))')

plt.show()
```

## 3. Площадь под многочленом.

1. Построить график подинтегральной функции.
2. Вычислить площадь под кривой на заданном отрезке как значение
определенного интеграла.
3. Закрасить область под графиком, чтобы визуализировать интеграл.

Вычислите интеграл:

$f(x) = 3x^4 - 2x^3 + x^2 - 5$

на интервале [−1, 1].

Дополнительные требования:
1. В каждой задаче используйте функцию fill_between() для закрашивания площади под кривой.
2. Добавьте подписи осей и заголовок.
3. При необходимости, отобразите оси X и Y для лучшей визуализации

```python
from scipy.integrate import quad

def f(x):
    return 3*x**4 - 2*x**3 + x**2 - 5

x = np.linspace(-1.3, 1.3, 500)
y = f(x)

integral, error = quad(f, -1, 1)

plt.figure(figsize=(8, 7))
plt.plot(x, y, label=r"$f(x) = 3x^4 - 2x^3 + x^2 - 5$", color='b')

x_fill = np.linspace(-1, 1, 400)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, color='black', alpha=0.5, label=f"Площадь ≈ {integral:.2f}")

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Площадь под графиком функции")
plt.legend()
plt.grid()

plt.show()
```

## 4. Волнообразная поверхность.

Визуализируйте функцию: $f(x, y) = sin(x^2 + y^2)$ на $x, y ∈ [−2π, 2π]$.


1. Построить трехмерный график функции f(x, y) в заданных пределах.
2. Использовать библиотеку Matplotlib для визуализации.
3. Оформить график: добавить заголовок, подписи осей и цветовую карту (если
уместно).

Дополнительные требования:
1. Используйте plot_surface() для плотных поверхностей.
2. Используйте plot_wireframe() для каркасных моделей.
3. Настроите цветовую карту через cmap .
4. Добавьте подписи осей и заголовки.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 50)
y = np.linspace(-2*np.pi, 2*np.pi, 50)
X, Y = np.meshgrid(x, y)

Z = np.sin(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title(r"Волнообразная поверхность: $f(x, y) = \sin(x^2 + y^2)$")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f(X, Y)")

fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```
